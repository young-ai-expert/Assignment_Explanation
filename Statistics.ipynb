# MEDIAN

1. Sort the data in ascending order.
2. We determine if the total no. of values are odd or even.
3. If it is odd, then we take the center value as the median.

    FOR EG: 
    0,2,3,2,1,0,0,2,0
    
    0,0,0,0,1,2,2,2,3
    So 1 is the median here.
    
4. If it is even, then we take the average of the two center values.

- Median is less suspectible to outliers when compared to mean, but it is harder to 
  calculate Median for a large no. of values say (10000 random values)

# Variance (σ<sup>2</sup>)

- Measures how spread out the data points are from the other samples.
- More variance usually means, more outliers.
- Variability describes how far apart data points lie from each other and from the center of a distribution.
<br>

Calculating Variance :
1. To calculate the average/mean of all data points.
2. Is to get the difference from the mean for all data points.
3. Square the each difference. (To remove  negative values)
4. Take the average of the squared differences and that is variance.




- Comparing the variance of multiple samples helps in accessing the group differences.
FOR EG : 


```python
l = [1,4,5,4,8]
mean = sum(l)/len(l)
```

```python
mean
```

```python
diff = [round(i-mean,2) for i in l]
```

```python
diff
```

```python
square_diff = [round(i**2,2) for i in diff]
```

```python
square_diff
```

```python
variance = round(sum(square_diff)/len(square_diff),2)
```

```python
variance
```

# Standard Deviation (σ) :

Calculating Standard Deviation:
1. Calculate Variance using the above steps.
2. Take square root of your variance, that is the standard deviation. (+ve square root)

<br>

- It is used as a way to identify outliers/data points that lie more than one standard
  deviation on either left or right side of the mean/
  
- You can tell how much an outlier is deviated from the maen by calculating how many
  standard deviation it is away from the mean.

```python
std_dev = round(variance**0.5,2)  #square root
```

```python
std_dev
```

```python
mean
```

```python
mean-std_dev
```

```python
round(mean+std_dev,2)
```
